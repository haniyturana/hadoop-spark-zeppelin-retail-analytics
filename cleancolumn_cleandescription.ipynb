```python
import pandas as pd


filepath = r"C:\Users\PC 10\Downloads\onlineretail_ii_1.parquet"
df = pd.read_parquet(filepath)
df.columns
```

```python
df = df.rename(columns={"Customer ID": "Customer_ID"})
```

```python
df.columns
```

```python
# fix missing value issue first
print(df[['Description', 'StockCode']].isnull().sum())
print("-" * 50)

pattern = r'(?i)(check|damaged|put aside|wrong|sold as|test|broken|adjust)'
df['Description'] = df['Description'].astype(str).str.strip()
df.loc[
    (df['Description'] == "") | 
    (df['Description'] == "?") | 
    (df['Description'].str.contains(pattern, na=False, regex=True)), 
    'Description'
] = None
```

```python
df['Description'] = df['Description'].fillna(df.groupby('StockCode')['Description'].transform('max'))

df['Description'] = df['Description'].fillna("NA")
print(df[['Description', 'StockCode']].isnull().sum())
```

```python
df.to_parquet(filepath, index=False)
```
